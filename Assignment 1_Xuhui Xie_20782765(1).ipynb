# transfer to DataFrame

```python
def transfer_data():
    
    columns = np.append(iris.feature_names,['target'])
    data = np.append(iris.data, iris.target[:,None], axis=1)
    index = pd.RangeIndex(start=1, stop=len(iris.data)+1, step=1)

    return pd.DataFrame(columns=columns, data=data, index=index)
```

```python
iris_df=transfer_data()
iris_df
```

# Question 2: KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

#1. divide the data into train, validation, and test sets
X_train2, X_test2, y_train2, y_test2 = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
X_train2, X_cv2, y_train2, y_cv2 = train_test_split(X_train2, y_train2, test_size=0.25, random_state=42)

#create claasifier object
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
#2. Train the classifier using the training data
knn.fit(X_train2, y_train2)
#Estimate the accuracy of the classifier using the test data
Accu_k_5 = knn.score(X_test2, y_test2)
Accu_k_5
```

```python
#3. find k for KNN using the cross validation data
klist = [1,5,10,15,20,25,30,35]
ka=np.zeros((8,2))

for x in range(len(klist)):
    knn = KNeighborsClassifier(n_neighbors=klist[x])
    knn.fit(X_train2, y_train2)
    ka[x,0] = klist[x]
    ka[x,1] = knn.score(X_cv2, y_cv2)

ka
```

```python
import matplotlib.pyplot as plt

plt.title('accuracy-k')
plt.xlabel('k')
plt.ylabel('accuracy')
plt.plot(ka[:,0], ka[:,1])
plt.xticks([0,5,10,15,20,25,30,35]);
```

```python
#4. using the best found parameters, train the model using the train set and test the model on the test set.
bestknn = KNeighborsClassifier(n_neighbors=25).fit(X_train2, y_train2)
print('Accuracy of KNN classifier on training set: {:.5f}'
     .format(bestknn.score(X_train2, y_train2)))
print('Accuracy of KNN SVC classifier on test set: {:.5f}'
     .format(bestknn.score(X_test2, y_test2)))
```

According to the result of the model on cross validation set using different k, the best k parameter for the KNN classifier is 25, and the accuracy of the model is 93.33% on the training set and is 100% on the test set.

# Question 3: SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X_train3, X_test3, y_train3, y_test3 = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import cross_val_score

Clist = [0.1, 0.5, 1, 2, 5, 10, 20, 50]
meanscores = np.zeros((8,2))

for cnum in range(len(Clist)):
    SVMclf = SVC(kernel='linear', C=Clist[cnum]).fit(X_train3, y_train3)
    scores = cross_val_score(SVMclf, X_train3, y_train3, cv=10)
    meanscores[cnum,0] = Clist[cnum]
    meanscores[cnum,1] = np.mean(scores)

meanscores
```

```python
import matplotlib.pyplot as plt

plt.title('mean accuracy-C')
plt.xlabel('C')
plt.ylabel('mean accuracy')
plt.plot(meanscores[:,0], meanscores[:,1]);
```

```python
svm_clf = SVC(kernel='linear', C=5, max_iter=5000).fit(X_train3, y_train3)
print('Coefficients:\n', svm_clf.coef_)
print('Intercepts:\n', svm_clf.intercept_)
print('Accuracy of Linear SVM classifier on training set: {:.5f}'
     .format(svm_clf.score(X_train3, y_train3)))
print('Accuracy of Linear SVM classifier on test set: {:.5f}'
     .format(svm_clf.score(X_test3, y_test3)))
```

According to the result of the model on cross validation set using different C, the best C parameter for the SVM classifier is 5, and the accuracy of the model is 97.5% on the training set and is 96.67% on the test set.

# Question 4: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

X_train4, X_test4, y_train4, y_test4 = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
DTdepth = [3, 5, 10, None]
DTmeanscores = np.zeros((4,2))

for d in range(len(DTdepth)):
    DTclf = DecisionTreeClassifier(max_depth = DTdepth[d])
    DTscores = cross_val_score(DTclf, X_train4, y_train4, cv=10)
    DTmeanscores[d,0] = DTdepth[d]
    DTmeanscores[d,1] = np.mean(DTscores)

DTmeanscores
```

```python
import matplotlib.pyplot as plt

plt.title('mean accuracy-depth')
plt.xlabel('depth')
plt.ylabel('mean accuracy')
plt.plot(DTmeanscores[:,0], DTmeanscores[:,1]);
```

```python
dt_clf = DecisionTreeClassifier(max_depth=3).fit(X_train4, y_train4)

print('Accuracy of Decision Tree classifier on training set: {:.5f}'
     .format(dt_clf.score(X_train4, y_train4)))
print('Accuracy of Decision Tree classifier on test set: {:.5f}'
     .format(dt_clf.score(X_test4, y_test4)))
```

According to the result of the model on cross validation set using different max_depth, the best max_depth parameter for the Decision Tree classifier is 3, and the accuracy of the model is 95.83% on the training set and is 100% on the test set.

# Question 4: Random Forest

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

X_train5, X_test5, y_train5, y_test5 = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
import numpy as np
import pandas as pd
RFtrees = {0:5, 1:10, 2:50, 3:150, 4:200}
RFdepth = {0:3, 1:5, 2:10, 3:None}
RFmatrix = np.mat(np.zeros((len(RFtrees), len(RFdepth))))
print(np.shape(RFmatrix))
RFmatrix
```

```python
for t in RFtrees.keys():
    for d in RFdepth.keys():
        RFclf = RandomForestClassifier(n_estimators=RFtrees[t], max_depth=RFdepth[d])
        RFscores = cross_val_score(RFclf, X_train5, y_train5, cv=10)
        RFmatrix[t,d] = np.mean(RFscores)
RFmatrix
```

```python
RFmeanscores = pd.DataFrame(RFmatrix,columns=RFdepth.values(),index=RFtrees.values())
RFmeanscores
```

```python
import seaborn as sns
sns.set()
ax = sns.heatmap(RFmeanscores, square=True, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel('depth')
plt.ylabel('trees');
```

```python
rf_clf = RandomForestClassifier(n_estimators=200, max_depth=3).fit(X_train5, y_train5)

print('Accuracy of Random Forest classifier on training set: {:.5f}'
     .format(rf_clf.score(X_train5, y_train5)))
print('Accuracy of Random Forest classifier on test set: {:.5f}'
     .format(rf_clf.score(X_test5, y_test5)))
```

According to the result of the model on cross validation set using different max_depth and number of trees, the best max_depth parameter is 3, the best number of trees parameter is 200, and the accuracy of the model is 95.83% on the training set and is 100% on the test set.

# Question 4: Gradient Boosting Decision Tree

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

X_train6, X_test6, y_train6, y_test6 = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.title('mean accuracy-nestimators')
plt.xlabel('nestimators')
plt.ylabel('mean accuracy')
plt.plot(GBDTmeanscores[:,0], GBDTmeanscores[:,1]);
```

```python
gb_clf = GradientBoostingClassifier(n_estimators=10).fit(X_train6, y_train6)

print('Accuracy of Gradient Boosting classifier on training set: {:.5f}'
     .format(gb_clf.score(X_train6, y_train6)))
print('Accuracy of Gradient Boosting classifier on test set: {:.5f}'
     .format(gb_clf.score(X_test6, y_test6)))
```

# Question 5: Analysis

1. Explain why you had to split the dataset into train and test sets?

Even if a learning algorithm fits a training set well, that does not mean it is a good hypothesis. It could overfit(high variance) or underfit(high bias), and as a result the predictions on new dataset would be poor. In other words, the error of your hypothesis as measured on the dataset with which you trained the parameters will be lower than the error on any other data set.

Thus, we need to split the dataset into training set and test set. The training set is the actual dataset that we use to train the model. The test set is used to evaluate the model we trained using training set, in order to provide an unbiased evaluation of a final model fit on the training dataset.

Given many models with different polynomial degrees, we can use a systematic approach to identify the best function. In order to choose the model of the hypothesis, you can test each degree of polynomial and look at the error result.

2. Explain why when finding the best parameters for KNN you didn't evaluate directly on the test set and had to use a validation test?

When you use test set to find the best parameters for KNN, the problem is this will not be a fair estimate of how well the hypothesis generalizes. The reason is what we've done is we have fit this extra parameter, that is k in KNN algorithm. If we fit this parameter k, using the test set, means that we chose the value of k that gave us the best possible performance on the test set. 

As a result, the performance of my parameter k on the test set is likely to be an overly optimistic estimate of generalization error. Because I had fit this parameter k to the test set, it is no longer fair to evaluate my hypothesis on this test set, because I fit my parameter k to this test set and chose the k using the test set. In other words, the hypothesis is likely to do better on this test set than it would on new examples that hasn't seen before.

Thus, we can not use test set to fit the parameters directly. Instead, validation set is used to fit the parameters in order to find the best parameter for this algorithm. Concretely, the validation set is used to to fine-tune the model hyperparameters.

3. What was the effect of changing k for KNN. Was the accuracy always affected the same way with an increase of k? Why do you think this happened?

When k=1, the one nearest neighbors classifier is over-fitting the training data in this case. It's trying to get correct predictions for every single training point while ignoring the general this trend between different classes. As k is increased, the result is a much smoother decision boundary, which represents a model with lower model complexity where the decision boundary has much less variance. If we increased k even higher to be the total number of points in the training set, the result would be a single decision region where all predictions would be the most frequent class in the training data.

According to the results, the testing accuracy didn't always increase the same way with an increases of k. It increases with k at the very beginning, reach the highest accuracy and then starts to go down as k increases. This is because when k is small, the model is much more complex with high variance and low bias which may cause overfitting; while k is large, the model becomes simple with high bias and low variance which may cause underfitting.

The purpose of machine learning is actually balance the bias-variance tradeoff, simultaneously minimize these two sources of error that prevent supervised learning algorithms from generalizing beyond their training set. As a result, we need to find some k in the middle which have the highest accuracy and lowest cost.

4. What was the relative effect of changing the max depths for decision tree, random forests, and gradient boosting? Explain the reason for this.

The max depth controls the maximum depth that the tree is allowed to grow to. The deeper it tree grows, the more complex the model will become.

For training accuracy, when increasing max_depth, the training accuracy will always improve and the training error will always go down. For testing accuracy, if the max_depth is set to be too high, then the decision tree might overfit the training data without capturing useful patterns as expected. This will lead to the high testing error. 

But if the max_depth is set too low, that is not good as well, this might result in the underfit problem of the model, namely giving the decision tree too little flexibility to capture the patterns and interactions in the training data. This will also cause the testing error to increase.


5. Comment on the effect of the number of estimators for Gradient Boosting and what was the relative effect performance of gradient boosting compared with random forest. Explain the reason for this.

In gradient boosting, the higher number of estimators will lead to higher number of trees as the number of trees grown in gradient boosting classifier is n classes mutiply by n estimators. which can generate less error and improve the model. Generally we want to have as many trees as possible in order to improve the model. More trees also mean more computational cost and after a certain number of trees, the improvement is negligible. Thus after sometime there is no significant improvement in error rate even if we are increasing no of tree.

Unlike the random forest method that builds and combines a forest of randomly different trees in parallel, the key idea of gradient boosting decision trees is that they build a series of trees. Where each tree is trained, so that it attempts to correct the mistakes of the previous tree in the series. In gradient boosting, the samples are weighted for sampling so that samples, which were predicted incorrectly get a higher weight and are therefore sampled more often. The idea behind this is that difficult cases should be emphasized during learning compared to easy cases. Because of this difference bagging can be easily paralleled, while boosting is performed sequentially.

Random forests and gradient boosting are all ensemble methods, the accuracy is the average over many trees created through bagging/boosting. Similarly, if you want to estimate an average of a real-valued random variable, you can take a sample. The expected variance will decrease as the square root of the sample size, and at a certain point the cost of collecting a larger sample will be higher than the benefit in accuracy obtained from such larger sample. That's why the influence of tree number on accuracy and error will converge as it increases.

6. What does the parameter C define in the SVM classifier? What effect did you observe and why do you think this happened?

The parameter C is the regularization parameter C parameter in SVM classifier defines the SVM optimization how much you want to avoid misclassifying each training example. It can be regarded as the reciprocal of lambda, which is the regularization parameter in linear regression and logistic regression.

For large values of C, the classifier will choose a smaller-margin hyperplane if that hyperplane does a better job of getting all the training points classified correctly. Conversely, a very small value of C will cause the classifier to look for a larger-margin separating hyperplane, even if that hyperplane misclassifies more points.

The reason for this is, in the cost function of SVM classifier, parameter C defines the slope of the boundary of the hyperplane. As a result, changing parameter C will influence the margin to the support vectors and how accurate these training examples will be classified.
